```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt # pandas uses matplotlib for plotting
```

A DataFrame is basically a table.  It is a 2D labelled data structure.  The columns can be different types.  You can think of it as a dict of Series objects (columns) if that helps.  Like a Series, a Dataframe has an index column.

We can build a pandas Dataframe in many ways, for example from a dict. The dict's keys become the column names and the dict's values become the column values.

We can use the pandas read_csv method to load data directly into a DataFrame. We will use the epl DataFrame in the following sections. This contains the football English Premier League results from the 2023-24 season as provided by ChatGPT!)

### Access columns in a Dataframe

We can access a columns or columns, col1, col2 in a DataFrame df in several ways:
* df['col1'] which returns a series
* df.col1 which returns a series (pandas allows us to refer to column names like a property! )
* df[['col1']] or df[['col1', 'col2']] which return DataFrames
* df.loc[ : , ['col1', 'col2']] which return DataFrames and is possibly best practice and most flexible

### Sort rows
We can sort by values or sort by index.  Sort by values is the more usual case.  

Order the rows in the dataframe by the number of games won (low to high), then in the case of any ties, by the number of games drawn.

```python
epl.sort_values(by = ['Won', 'Drawn'], ascending=True)
```

### Filter rows
Filter both rows and columns of the dataframe in various ways using the loc method and the index / column names.

```python
epl.loc['LIV', :] # one row, all columns, returns a Series
```

```python
epl.loc[['LIV'], :] # one row, all columns, returns a DataFrame
```

```python
epl.loc[['LIV', 'MCY'], :] #   two rows, all columns, returns a DataFrame
```

```python
epl.loc[['LIV', 'MCY'], ['Team', 'Won']] # two rows, two columns
```

### Filter Rows by Values

We have already seen that a boolean expression returns a Series of bools

```python
epl.Won >= 25
```

We can use this boolean expression to filter the rows of the DataFrame where the bools are True

```python
epl.loc[epl.Won >= 25, :]
```

### Drop Columns

We can remove columns from a DataFrame using the drop() method. By default, drop() returns a new DataFrame without modifying the original.

```python
# Drop a single column (returns new DataFrame, original unchanged)
epl_no_lost = epl.drop('Lost', axis=1)
epl_no_lost.head()
```

```python
# Drop multiple columns
epl_simplified = epl.drop(['Drawn', 'Lost'], axis=1)
epl_simplified.head()
```

### Rename Columns

We can rename one or more columns using the rename() method. Pass a dictionary mapping old names to new names.

```python
# Rename single or multiple columns
epl_renamed = epl.rename(columns={'GF': 'Goals_For', 'GA': 'Goals_Against'})
epl_renamed.head()
```

### Reset Index

We can convert the index back to a regular column using reset_index(). This is useful when you want to make the index a column again or switch to the default numeric index.

```python
# Reset index to default numeric index (Code becomes a regular column)
epl_reset = epl.reset_index()
epl_reset.head()
```

### Using query() for Filtering

The query() method provides an alternative, more readable way to filter DataFrames using SQL-like string expressions. This can be cleaner than using boolean indexing with .loc[].

```python
# Simple query - teams that won 25 or more games
epl.query('Won >= 25')
```

```python
# Complex query with multiple conditions - teams with more than 20 wins AND more than 70 points
epl.query('Won > 20 and Points > 70')
```
